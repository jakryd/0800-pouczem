### Flow control 

#### Conditional expressions: `if`, `elif`, `else`

If 2**2 is equal to 4 print 'Obvious!'

```python
if 2**2 == 4:
    print('Obvious!')
```

You can check additional conditions using `elif` or `else` keywords:

```python
if 2**3 == 4:
    print('It is equal to 4!')
elif 2**3 == 8:
    print('It is equal to 8!')
else:
    print('I do not have an idea what is going on!')
```

Conditional expressions can also be used with list comprehension:

```python
l = range(10)
even_squares = [x ** 2 for x in l if x % 2 == 0]
print(even_squares)
```

Also with `else`:

```python
l = range(10)
even_squares = [x ** 2 if x % 2 == 0 else x ** 3 for x in l]
print(even_squares)
```

In general, 
`[f(x) if condition else g(x) for x in sequence]`

And, for list comprehensions with if conditions only,
`[f(x) for x in sequence if condition]`

#### Loops: `for` with `range`

Interate in a range [0, 4):

```python
for i in range(4):
    print(i)
```

In a range [2,4):

```python
for i in range(2, 4):
    print(i)
```

In a range [1, 6) every second item:

```python
for i in range(1, 6, 2):
    print(i)
```

#### Loops: `while` with `break`, `continue`

```python
i = 0.1
while i < 10:
    print(i)
    i += 0.5
```

```python
for i in range(0, 10, 2):
    print(i)
```

```python
n = 10
i = 0.0
while i < n:
    if i == 5.0:
        break
    print(i)
    i += 1
```

```python
z = 1 + 1j
while abs(z) < 100:
    z = z**2 + 1
    print(z)
```

Stop iteration if the imaginary part of z is equal to 4:

```python
z = 1 + 1j
while abs(z) < 100:
    if z.imag == 4:
        break
    z = z**2 + 1
    print(z)
```

Go to the next iteration if `element` is 0:

```python
a = [1, 0, 2, 4]
for element in a:
    if element == 0:
        continue
    print(1. / element)
```

**Getting indexes**

You can retrive the iteration index in two ways:

```python
words = ('cool', 'powerful', 'readable')
for i in range(0, len(words)):
    print((i, words[i]))
```

```python
for index, item in enumerate(words):
    print((index, item))
```

Iterating over a dictionary is also possible:

```python
d = {'a': 1, 'b':1.2, 'c':1j}
```

```python
for key, val in d.items():
    print(key, val)
```

Compute the decimals of Pi using the Wallis formula:

$\pi = 2 \prod_{i=1}^{\infty} \frac{4i^2}{4i^2 - 1}$

```python
idx = []
errors = []
```

```python
pi = 2.0
for i in range(1, 1000):
    sqr = 4.0 * i ** 2
    pi *= sqr / (sqr - 1)
    error = abs(math.pi - pi)

    idx.append(i)
    errors.append(error)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(idx, errors)
```
