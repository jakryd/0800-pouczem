### Basics of Python

Python is a high-level, dynamically typed multiparadigm programming language. Python code is often said to be almost like pseudocode, since it allows you to express very powerful ideas in very few lines of code while being very readable. As an example, here is an implementation of the classic quicksort algorithm in Python:

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
```

Another example -- calculating the Fibbonacci sequence:

```python
def fibbonacci(n):
    if n == 0 or n == 1: return 1
    else: return n + fibbonacci(n-1)
    
print(fibbonacci(6))
```

#### Strings

You can use single quotes `' '` or double quotes `" "` in string literals:

```python
hello = 'hello'
world = "world"
print(hello, len(hello))
```

And concatenate then like this:

```python
hw = hello + ' ' + world
print(hw)
```

You can format strings using `format` method:

```python
hw12 = '{} {} {}'.format(hello, world, 12)
print(hw12)
```

String objects have a lot of useful methods; for example:

Capitalize a string:

```python
s = "hello"
print(s.capitalize())
```

Convert a string to uppercase:

```python
print(s.upper())
```

Right-justify a string, padding with spaces:

```python
print(s.rjust(7))
```

Center a string, padding with spaces:

```python
print(s.center(7))
```

Replace all instances of one substring with another

```python
print(s.replace('l', '(ell)'))
```

Strip leading and trailing whitespace

```python
print('  world '.strip())
```

You can find a list of all string methods in the [documentation](https://docs.python.org/3.7/library/stdtypes.html#string-methods).
